```python
# Task 2: Mark the success/failed launches for each site on the map
import pandas as pd
import folium
from folium.plugins import MarkerCluster
from folium.features import DivIcon

# Step 1: Load the dataset
spacex_df = pd.read_csv('spacex_launch_geo.csv')

# Step 2: Select relevant columns and get unique launch sites
launch_sites_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']].groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]

# Step 3: Create a map centered around NASA JSC
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Add a marker for NASA JSC
circle = folium.Circle(nasa_coordinate, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('NASA Johnson Space Center'))
marker = folium.map.Marker(
    nasa_coordinate,
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'NASA JSC',
        )
    )
site_map.add_child(circle)
site_map.add_child(marker)

# Display the map (we will add other markers in the next steps)
site_map

```

```python
# Step 4: Add markers for each launch site to the map
for index, site in launch_sites_df.iterrows():
    folium.CircleMarker(
        location=[site['Lat'], site['Long']],
        radius=5,
        color='blue',
        popup=site['Launch Site'],
        fill=True,
        fill_color='blue',
        fill_opacity=1.0,
    ).add_to(site_map)

# Step 5: Group the data by launch site and class to get the count of successes and failures
site_success_failure = spacex_df.groupby(['Launch Site', 'class']).size().reset_index(name='Count')

# Step 6: Add success and failure circles to the map for each launch site
for index, site in launch_sites_df.iterrows():
    site_name = site['Launch Site']
    site_data = site_success_failure[site_success_failure['Launch Site'] == site_name]

    for _, record in site_data.iterrows():
        if record['class'] == 1:
            color = 'green'
        else:
            color = 'red'
        folium.Circle(
            location=[site['Lat'], site['Long']],
            radius=record['Count']*50,
            color=color,
            fill=True,
            fill_color=color,
        ).add_to(site_map)

# Display the map with all the markers
site_map
```

```python
from folium.plugins import MarkerCluster

# Create a MarkerCluster object and add it to the site_map
marker_cluster = MarkerCluster().add_to(site_map)

# Loop through the dataframe and add each data point to the marker cluster
for lat, lng, label in zip(spacex_df.Lat, spacex_df.Long, spacex_df['Launch Site']):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(marker_cluster)

# Display the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
from folium.plugins import MousePosition

# Add Mouse Position to the map to get the coordinate (Lat, Long) for a mouse over a point on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)

# Assuming you have selected the following coordinates for the closest coastline point (replace with the actual values)
coastline_lat = 28.56367  
coastline_lon = -80.57163  

# Assuming you are calculating distance to this launch site (replace with the actual values)
launch_site_lat = 28.5721  
launch_site_lon = -80.6480  

# Calculate the distance between the launch site and the coastline point
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Create and add a marker on the closest coastline point
distance_marker = folium.Marker(
   [coastline_lat, coastline_lon],
   icon=DivIcon(
       icon_size=(20,20),
       icon_anchor=(0,0),
       html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
       )
   )
site_map.add_child(distance_marker)

# Draw a line between the launch site and the coastline point
lines = folium.PolyLine(locations=[[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]], weight=1)
site_map.add_child(lines)

# Display the map
site_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
# Define the coordinates of the launch site (replace with actual values)
launch_site_lat = 28.5721  
launch_site_lon = -80.6480  

# Coordinates of important points
points = {
    "Railway Line": [28.57212, -80.58528],
    "Road": [28.52474, -80.63176],
    "Nearby City": [29.02489, -80.92703]
}

# Loop through the points and add distance markers and lines to the map
for point_name, (lat, lon) in points.items():
    # Calculate the distance between the launch site and the point
    distance = calculate_distance(launch_site_lat, launch_site_lon, lat, lon)
    
    # Create and add a marker at the point with a popup displaying the distance from the launch site
    distance_marker = folium.Marker(
       [lat, lon],
       icon=DivIcon(
           icon_size=(20,20),
           icon_anchor=(0,0),
           html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
           )
       )
    site_map.add_child(distance_marker)
    
    # Create and add a line between the launch site and the point
    line = folium.PolyLine(locations=[[launch_site_lat, launch_site_lon], [lat, lon]], weight=1)
    site_map.add_child(line)

# Display the updated map
site_map
```
